```python
# Importing Required libraries
import pandas as pd
import numpy as np                   
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
# Loading dataset
house_df = pd.read_csv("/content/train.csv")
house_df
```

```python
house_df = house_df.set_index(["Id"])
house_df
```

```python
all_data_na = (house_df.isnull().sum() / len(house_df)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio (in %)':all_data_na})
missing_data.head(20)
```

```python
house_df = house_df.drop(['PoolQC', 'MiscFeature','Alley','Fence','FireplaceQu'], axis=1)
house_df
```

```python
a = house_df.isnull().sum().sort_values(ascending=False)[:19].to_frame()
l = list(a.index.values)
house_df[l].info()
```

```python
num_columns = list(house_df[l].select_dtypes(include=np.number).columns)
cat_columns = list(house_df[l].select_dtypes(exclude=np.number).columns)

print("Numeric Variables : ",num_columns)
print()
print("Categorical Variables : ",cat_columns)
```

```python
for i in list(house_df[l].columns):
    if i in num_columns:
        house_df[i].fillna(house_df[i].median(), inplace=True)
    else:
        house_df[i].fillna(house_df[i].mode()[0], inplace=True)
```

```python
all_data_na = (house_df.isnull().sum() / len(house_df)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)[:30]
missing_data = pd.DataFrame({'Missing Ratio (in %)':all_data_na})
missing_data
```

```python
house_df
```

```python
num_columns = list(house_df.select_dtypes(include=np.number).columns)
numeric_house_df = house_df[num_columns]
numeric_house_df
```

```python
#LOG SCALING

def log_scaling(df):
    new = df
    for i in list(new.columns):
        new[i] = np.log(df[i])
    return new

california_log_scaling = log_scaling(numeric_house_df)
california_log_scaling
```

```python
# Min-Max
def min_max(df):
    new = df
    for i in list(new.columns):
        new[i] = (new[i] - min(new[i]))/(max(new[i]) - min(new[i]))
    return new

california_min_max_udf = min_max(numeric_house_df)
california_min_max_udf
```

```python
# validating Min-Max
min_max_scaler = MinMaxScaler()

california_min_max_sklearn = pd.DataFrame()

l = list(numeric_house_df.columns)

california_min_max_sklearn[l] = min_max_scaler.fit_transform(numeric_house_df[l])

california_min_max_sklearn
```

```python
# Z-Score
def z_scores(df):
    new = df
    for i in list(new.columns):
        new[i] = (new[i] - np.mean(new[i]))/(np.std(new[i]))
    return new

california_z_score_udf = z_scores(numeric_house_df)
california_z_score_udf
```

```python
# Validating z-score
z_score = StandardScaler()

california_z_scores_sklearn = pd.DataFrame()

l = list(numeric_house_df.columns)

california_z_scores_sklearn[l] = z_score.fit_transform(numeric_house_df[l])

california_z_scores_sklearn
```
