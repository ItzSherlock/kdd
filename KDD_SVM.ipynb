**Aim :** To implement SVM using python on the given dataset(Fish.csv)
 

```python
# Importing required libraries

import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Fish.csv")
data
```

```python
# Define the Feature and the Target variables
X = data.drop(['Species'], axis = 'columns')
y = data.Species
```

```python
# Split the data into train, test sets using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
```

```python
# Instantiate Linear SVC object and train using training data
model = SVC(kernel = 'linear', C = 1)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy : {acc*100}%")
```

```python
# generate classification report
cr = classification_report(y_test, y_pred)
print(cr)
```
